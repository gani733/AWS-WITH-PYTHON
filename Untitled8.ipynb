```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
path="/content/RTA Dataset.csv"
df=pd.read_csv(path)
df.head()
```

```python
path="/content/cleaned.csv"
df=pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df['Accident_severity'].value_counts())
df['Accident_severity'].value_counts().plot(kind='bar')
```

```python
df['Educational_level'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(12,10))
sns.catplot(data=df,x='Educational_level',y='Accident_severity')
plt.xlabel('Educational_level')
plt.xticks(rotation=60)
plt.show()
```

```python
print(df['Road_surface_type'].value_counts())
plt.figure(figsize=(12,10))
sns.countplot(x='Road_surface_type',hue='Accident_severity',data=df)
plt.xlabel('Road_surface_type')
plt.xticks(rotation=60)
plt.show()
```

```python
print(df['Road_surface_conditions'].value_counts())
plt.figure(figsize=(12,10))
sns.countplot(x='Road_surface_conditions',hue='Accident_severity',data=df)
plt.xlabel('Rode condition type')
plt.xticks(rotation=60)
plt.show()
```

```python
print(df['Road_Surface_Conditions'].value_counts())
plt.figure(figsize=(12,10))
sns.countplot(x='Road_Surface_Conditions',hue='Accident_severity',data=df)
plt.xlabel('Rode condition type')
plt.xticks(rotation=60)
plt.show()
```
